```python
%pylab
%matplotlib inline
import theory as th
```

```python
def FindGlobSize(eps, zvar):
    N = len(zvar)
    tmpList = [v for v in zvar if v>eps]
    return len(tmpList)
```

```python
T = linspace(0.1, 200, 200)
globSize = []
for x in T:
    zvar = th.var_z_1D(500, x)
    globSize.append(FindGlobSize(1e-1, zvar))
globSize = array(globSize)
plot(T, globSize)
xlabel(r'$1/F$')
ylabel('Effective glob size')
```

```python
tau = globSize ** 2 / (3*pi*pi)
plot(T, tau)
xlabel(r'$1/F$')
ylabel('Relaxation time')
```

```python
%matplotlib inline
from sympy import *
init_printing()
x, x0, a, b, c, t, L, D = symbols('x x0 a b c t L D')
sigma, mu = symbols('sigma mu')
k, m, n = symbols('k m n', integer=True)
f, g, h, p, q = symbols('f g h p q', cls=Function)
```

The solution of single drifting Brownian partile in refleting boundaries $[0,L]$ can be written as 

```python
f1 = 1/(sigma*sqrt(2*pi*t))*exp(-2*mu*n*L/sigma**2)*exp(-(x+2*n*L-x0-mu*t)**2/(2*sigma**2*t))
f2 = 1/(sigma*sqrt(2*pi*t))*exp(-2*mu*(n*L+x0)/sigma**2)*exp(-(x+2*n*L+x0-mu*t)**2/(2*sigma**2*t))
g = 1/sqrt(2*pi)*integrate(exp(-a**2/2), (a, -oo, x))
f3 = 2*mu/sigma**2*(exp(2*mu*(x+n*L)/sigma**2))*(1-g.subs(x, (x+2*n*L+x0+mu*t)/(sigma*sqrt(t))))
f4 = 2*mu/sigma**2*(exp(2*mu*(x-(n+1)*L)/sigma**2))*(1-g.subs(x, (x-2*(n+1)*L+x0+mu*t)/(sigma*sqrt(t))))
p = Sum(f1, (n, -oo, +oo)) + Sum(f2, (n, -oo, +oo)) - Sum(f3, (n, 0, +oo)) + Sum(f4, (n, 0, +oo))
p
```

```python
(sigma**2*diff(p, x)/2-mu*p).subs(x, 0).doit()
```

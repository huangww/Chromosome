```python
%cd '~/Work/chromosome'
%pylab
%matplotlib inline
import statsmodels.tsa.stattools as ss
```

```python
T = 100
fname = 'data/rg1D_N100_T' + str(T) + '_eq.dat'
data = loadtxt(fname)
x = data[:, 1]
t = data[:, 0]
plot(t, x,'-')
# xlim(0, 1e5)
xlabel(r'$t$')
ylabel(r'$<Rg(t)>$')
lags = int(1e5)
acf= ss.acf(x, nlags=lags, fft=True)
end = next(i for i,v in enumerate(acf) if v<0)
t = t[:end]
acf = acf[:end]
# fname = 'data/acf1D_N100_T' + str(T) + '.dat'
# savetxt(fname, transpose([t[:len(acf)], acf]))
```

```python
plot(t, acf,'-')
yscale('log')
xlabel(r'$\delta t$')
ylabel(r'$ACF$')
ylim(1e-3, 1)
# t0 = int(len(t)/10)
# t1 = int(len(t)/2)
t0 = 500
t1 = 5000
tFit = t[t0:t1]
acfFit = acf[t0:t1]
coefs = polyfit(tFit, log(acfFit), 1)
tau = -1./coefs[0]
print 'tau =', tau
plot(tFit, exp(coefs[0]*tFit+coefs[1]))
```

```python
print T, tau
f = open('data/tauForce.dat', 'ab')
savetxt(f, (T, tau), newline=' ')
f.write('\n')
f.close()
```

# 1D ASEP Results

## $\tilde{T} = 1$, strong external force field 

### Relaxation from random initial configuration

```python
data = loadtxt('data/rg1D_N100_T1.dat')
plot(data[:,0], data[:,1])
xlabel(r'$t$')
ylabel(r'$<Rg(t)>$')
xlim(0, 1000)
```

#### Normalized gyration radius

```python
x = data[:,1]
xeq = x[-len(x)/10:].mean()
x = fabs(x - xeq)
x = x / x.max()
t = data[:,0]
plot(t, x)
xlabel(r'$t$')
ylabel(r'$|<Rg(t)>-<Rg^{eq}>|$')
xlim(0, 1000)
yscale('log')
# xscale('log')
```

### Auto-Correlation Fuction (ACF) of equilibrium sample

```python
data = loadtxt('data/rg1D_N100_T1_eq.dat')
x = data[:, 1]
t = data[:, 0]
plot(t, x,'-')
xlim(0, 1e4)
xlabel(r'$t$')
ylabel(r'$<Rg(t)>$')
```

```python
lags = int(2e3)
acf= ss.acf(x, nlags=lags, fft=True)
plot(t[:lags+1], acf,'-')

# acf = loadtxt('data/acf1D_N100_T100.dat')[:,1]
# plot(t[:len(acf)], acf)

yscale('log')
xlabel(r'$\delta t$')
ylabel(r'$ACF$')
# xlim(0, 2e2)
ylim(1e-3, 1)

t0 = int(0)
t1 = int(1e3)
tFit = t[t0:t1]
acfFit = acf[t0:t1]
coefs = polyfit(tFit, log(acfFit), 1)
print 'tau =', -1./coefs[0]
plot(tFit, exp(coefs[0]*tFit+coefs[1]))
```

# Comparision of 1D ASEP and 3D BD

## $\tilde{T}=1$, strong external force field

```python
data1d = loadtxt('data/rg1D_N100_T1.dat')
data3d = loadtxt('data/rgmean_N100_T1.dat')
t1 = data1d[:,0]
x1 = data1d[:,1]
x1 = fabs(x1 - x1[-len(x1)/10:].mean())
x1 = x1 / x1.max()
tscale = 4.5
t3 = arange(len(data3d))*tscale
x3 = data3d[:,3]
x3 = fabs(x3 - x3[-len(x3)/10:].mean())
x3 = x3 / x3.max()
plot(t1, x1, 'b-')
plot(t3, x3, 'r-')
yscale('log')
xlim(0,1000)
xlabel(r'$t$')
ylabel(r'$|<Rg(t)>-<Rg^{eq}>|$')
```

```python
data1d = loadtxt('data/rg1D_N100_T1_eq.dat')
data3d = loadtxt('data/rg_N100_T1_0.dat')
t1 = data1d[:,0]
x1 = data1d[:,1]
tscale = 4.5
t3 = data3d[:,0]
x3 = data3d[:,1]
t0 = 0
t1 = t1[t0:] - t1[t0]
x1 = x1[t0:]
acf1 = ss.acf(x1, nlags=len(x1), fft=True)
plot(t1, acf1,'b-')
t0 = 1000
t3 = t3[t0:] - t3[t0]
x3 = x3[t0:]
acf3 = ss.acf(x3, nlags=len(x3), fft=True)
plot(t3, acf3, 'r-')
xlim(0, 1000)
yscale('log')
xlabel(r'$\delta t$')
ylabel(r'$ACF$')
```

# 1D ASEP Results

## $\tilde{T} = 1$, strong external force field 

### Relaxation from random initial configuration

```python
data = loadtxt('data/rg1D_N100_T1.dat')
plot(data[:,0], data[:,1])
xlabel(r'$t$')
ylabel(r'$<Rg(t)>$')
```

#### Normalized gyration radius

```python
x = data[:,1]
xeq = x[-len(x)/10:].mean()
x = fabs(x - xeq)
x = x / x.max()
t = data[:,0]
plot(t, x)
xlabel(r'$t$')
ylabel(r'$|<Rg(t)>-<Rg^{eq}>|$')
# xlim(0, 1000)
yscale('log')
# xscale('log')
```

### Auto-Correlation Fuction (ACF) of equilibrium sample

```python
data = loadtxt('data/rg1D_N100_T1_eq.dat')
x = data[:, 1]
t = data[:, 0]
plot(t, x,'-')
xlim(0, 1000)
xlabel(r'$t$')
ylabel(r'$<Rg(t)>$')
```

```python
lags = 500
acf = ss.acf(x, nlags=lags, fft=True)
plot(t[:lags+1], acf,'-')
yscale('log')
xlabel(r'$\delta t$')
ylabel(r'$ACF$')
t0 = 200
tFit = t[:t0]
acfFit = acf[:t0]
coefs = polyfit(tFit, log(acfFit), 1)
print 'tau =', -1./coefs[0]
plot(tFit, exp(coefs[0]*tFit+coefs[1]))
```

# 3D Brownian Dynamics Simulation 

## $\tilde{T} = 1$, strong external force field 

### Relaxation from random initial configuration

```python
data = loadtxt('data/rgmean_N100_T1.dat')
tscale = 1
t = arange(len(data))*tscale
rg = data[:,3]
plot(t, rg)
xlabel(r'$t$')
ylabel(r'$<Rg(t)>$')
```

#### Normalized gyration radius

```python
x = data[:,3]
xeq = x[-len(x)/10:].mean()
x = fabs(x - xeq)
x = x / x.max()
plot(t, x)
xlabel(r'$t$')
ylabel(r'$|<Rg(t)>-<Rg^{eq}>|$')
xlim(0, 1000)
# yscale('log')
```

### Auto-Correlation Fuction (ACF) of equilibrium sample

```python
data = loadtxt('data/')
x = data[t0:, 3]
t = arange(len(data))*tscale
t = t[t0:] - t[t0]
plot(t, x)
xlabel(r'$t$')
ylabel(r'$<Rg(t)>$')
```

```python
acf = ss.acf(x, nlags=len(x))
plot(t, acf)
xlim(0, 50)
# yscale('log')
xlabel(r'$\delta t$')
ylabel(r'$ACF$')
```

# Comparision of 1D ASEP and 3D BD

## $\tilde{T}=1$, strong external force field

```python
data1d = loadtxt('data/rg1D_N100_T1.dat')
data3d = loadtxt('data/rgmean_N100_T1.dat')
t1 = data1d[:,0]
x1 = data1d[:,1]
x1 = fabs(x1 - x1[-len(x1)/10:].mean())
x1 = x1 / x1.max()
tscale = 9.5
t3 = arange(len(data3d))*tscale
x3 = data3d[:,3]
x3 = fabs(x3 - x3[-len(x3)/10:].mean())
x3 = x3 / x3.max()
plot(t1, x1, 'b-')
plot(t3, x3, 'r-')
yscale('log')
xlim(0,1000)
xlabel(r'$t$')
ylabel(r'$|<Rg(t)>-<Rg^{eq}>|$')
```

```python
data1d = loadtxt('data/rg1D_N100_T1.dat')
data3d = loadtxt('data/rgmean_N100_T1.dat')
t1 = data1d[:,0]
x1 = data1d[:,1]
tscale = 10
t3 = arange(len(data3d))*tscale
x3 = data3d[:,3]
t0 = 600
t1 = t1[t0:] - t1[t0]
x1 = x1[t0:]
acf1 = ss.acf(x1, nlags=len(x1))
plot(t1, acf1,'b-')
t3 = t3[t0:] - t3[t0]
x3 = x3[t0:]
acf3 = ss.acf(x3, nlags=len(x3))
plot(t3, acf3, 'r-')
xlim(0, 1000)
# yscale('log')
xlabel(r'$\delta t$')
ylabel(r'$ACF$')
```

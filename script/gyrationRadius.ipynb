```python
%matplotlib inline
from sympy import *
init_printing()
```

```python
i,j,k,l,T,N,mu = symbols('i j k l T, N, mu')
```

```python
mean_x_i = 0
mean_y_i = 0
mean_z_i = T*(log((i-mu)/(T*sinh((i-mu)/T)))-log((-mu)/(T*sinh((-mu)/T))))
```

```python
plot(mean_z_i.subs({T:1, N:100, mu:50}),
     mean_z_i.subs({T:10, N:100, mu:50}),
     mean_z_i.subs({T:50, N:100, mu:50}),
     (i,0,100)) 
```

```python
mean_square_r_i = mean_z_i * mean_z_i
msri = mean_square_r_i
simplify(msri)
```

```python
sigma_square_z_kl = T*((coth((l-mu)/T)-T/(l-mu))-(coth((k-mu)/T)-T/(k-mu)))
var_z_i = sigma_square_z_kl.subs({k:0,l:i})*sigma_square_z_kl.subs({k:i,l:N})/sigma_square_z_kl.subs({k:0,l:N})
simplify(var_z_i)
```

```python
plot(var_z_i.subs({T:1, N:100, mu:50}),
     var_z_i.subs({T:10, N:100, mu:50}),
     var_z_i.subs({T:100, N:100, mu:50}),
     (i,0,100)) 
```

```python
mean_e_z = coth((mu-i)/T) - T/(mu-i)
var_e_z = (T/(mu-i))**2 - (1/sinh(mu-i)/T)**2
second_moment_xy = 1 - (var_e_z + mean_e_z * mean_e_z)
simplify(second_moment_xy)
```

```python
sigma_square_xy_kl = summation(second_moment_xy, (i, k, l))
```

```python
var_xy_i = sigma_square_xy_kl.subs({k:0,l:i})*sigma_square_xy_kl.subs({k:i,l:N})/sigma_square_xy_kl.subs({k:0,l:N})
```

```python
var_r_i = var_xy_i + var_z_i
```

```python
small = Min(i,j)
large = Max(i,j)
covar_xy_ij = sigma_square_xy_kl.subs({k:0,l:small})*sigma_square_xy_kl.subs({large:i,l:N})/sigma_square_xy_kl.subs({k:0,l:N})
covar_z_ij = sigma_square_z_kl.subs({k:0,l:small})*sigma_square_z_kl.subs({large:i,l:N})/sigma_square_z_kl.subs({k:0,l:N})
covar_r_ij = covar_xy_ij + covar_z_ij
```

```python
gyrationRadius = summation(var_r_i + msri, (i, 0, N))/N - summation(covar_r_ij - mean_z_i*mean_z_i.subs(i,j), (i, 0, N), (j, 0, N))/(N*N)
simplify(gyration)
```
